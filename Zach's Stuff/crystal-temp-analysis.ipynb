```python
def fit_function(l_f, l_sh, n_f, n_sh, d_33, L, P, S_eff, m, A):
    CONST = ((2*pi * d_33 * L * P)**2 / (epsilon_0 * c * n_sh * (n_f**2) * (l_sh)**2 * S_eff))
    beta_sh = 2*pi/l_sh
    beta_f  = 2*pi/l_f
    m_SCALED = (beta_sh - 2*beta_f + m*(2*pi/A)) * L/2
    def sinc(x): return jnp.sin(x)/x
    return CONST * sinc(m_SCALED)**2

# constants
l_sh = 780e-9
n_f = 1.7280
n_sh = 1.816
d_33 = 9.74e-12
w_0 = 1.5e-3
S_eff = w_0**2 * pi/2
A = 45.2e-6

def fit_function_2(x, L, P, m):
    return fit_function(x, l_sh, n_f, n_sh, d_33, L, P, S_eff, m, A)

# fit parameters
L = 0.03
P = 0.2
m = 0.1

x = jnp.linspace(1557.0e-9, 1557.25e-9, 1000)

plt.plot(x, fit_function_2(x, L, P, m))
```

```python
from scipy.optimize import curve_fit

fig, ax = plt.subplots(figsize=(12, 3), dpi=300)

ax.set_title('Phase matching condition')
ax.set_xlabel('Temperature ($^{\\circ}$C)')
ax.set_ylabel('Intensity (nW)')

# Define sinc^2 function
def sinc_squared(x, A, x0, w, offset):
    """
    Sinc squared function for phase matching
    A: amplitude
    x0: center position (peak temperature)
    w: width parameter
    offset: vertical offset
    """
    arg = (x - x0) / w
    sinc = jnp.sinc(arg)  # numpy sinc is sin(pi*x)/(pi*x)
    return A * sinc**2 + offset

# Prepare data (convert JAX arrays to numpy)
x_data = jnp.array(temperature_data[20:])
y_data = jnp.array(intensity_data[20:])

T_arr = jnp.linspace(x_data[0], x_data[-1], 1000)

for n in [4,3,2,1,0]:

    # Initial parameter guesses (can change these depending)
    initial_guess = [
        40,                       # A: peak height
        x_data[jnp.argmax(y_data)]+n*10,         # x0: center at peak
        5.0,                                  # w: width parameter
        0                                    # offset: baseline
    ]

    # Perform curve fitting
    popt, pcov = curve_fit(sinc_squared, x_data, y_data, p0=initial_guess)

    # Extract fitted parameters
    A_fit, x0_fit, w_fit, offset_fit = popt
    perr = jnp.sqrt(jnp.diag(pcov))  # Parameter uncertainties

    plt.plot(T_arr, sinc_squared(T_arr, *popt), color=[(4-n)/4, 0, n/4], alpha=(10-n)/10, label=f'Fringe {n+1}')

print("Fitted parameters:")
print(f"Amplitude (A): {A_fit:.3f}  {perr[0]:.3f} nW")
print(f"Center (x0): {x0_fit:.3f}  {perr[1]:.3f} C")
print(f"Width (w): {w_fit:.3f}  {perr[2]:.3f} C")
print(f"Offset: {offset_fit:.3f}  {perr[3]:.3f} nW")

# Calculate R-squared
residuals = y_data - sinc_squared(x_data, *popt)
ss_res = jnp.sum(residuals**2)
ss_tot = jnp.sum((y_data - jnp.mean(y_data))**2)
r_squared = 1 - (ss_res / ss_tot)
print(f"\nR = {r_squared:.4f}")

plt.scatter(x_data, y_data, color='k')
plt.legend()
```

```python
import jax
import jax.numpy as jnp
from jax import grad, jit, random
import optax

# Initialize random key
key = random.PRNGKey(42)

# Prepare data
x_data = temperature_data.reshape(-1, 1)  # Shape: (N, 1)
y_data = intensity_data.reshape(-1, 1)     # Shape: (N, 1)

# Normalize data for better training
x_mean, x_std = jnp.mean(x_data), jnp.std(x_data)
y_mean, y_std = jnp.mean(y_data), jnp.std(y_data)

x_norm = (x_data - x_mean) / x_std
y_norm = (y_data - y_mean) / y_std

# Initialize network parameters
def init_network_params(layer_sizes, key):
    """Initialize parameters for a 2-layer fully connected network"""
    params = []
    keys = random.split(key, len(layer_sizes))
    
    for i in range(len(layer_sizes) - 1):
        key = keys[i]
        w_key, b_key = random.split(key)
        scale = jnp.sqrt(2.0 / (layer_sizes[i] + layer_sizes[i+1]))
        W = random.normal(w_key, (layer_sizes[i], layer_sizes[i+1])) * scale
        b = jnp.zeros(layer_sizes[i+1])
        params.append({'W': W, 'b': b})
    
    return params

# Define network architecture
layer_sizes = [1, 32, 32, 1]  # Input -> Hidden1 -> Hidden2 -> Output
params = init_network_params(layer_sizes, key)

# Forward pass
def forward(params, x):
    """Forward pass through the network"""
    activation = x
    
    # First layer with ReLU
    activation = jnp.dot(activation, params[0]['W']) + params[0]['b']
    activation = jax.nn.relu(activation)
    
    # Second layer with ReLU
    activation = jnp.dot(activation, params[1]['W']) + params[1]['b']
    activation = jax.nn.relu(activation)
    
    # Output layer (linear)
    output = jnp.dot(activation, params[2]['W']) + params[2]['b']
    
    return sinc_squared(x, *popt)

# Loss function (MSE)
def loss_fn(params, x, y):
    """Mean squared error loss"""
    predictions = forward(params, x)
    return jnp.mean((predictions - y) ** 2)

# Training loop
optimizer = optax.adam(learning_rate=0.001)
opt_state = optimizer.init(params)

# JIT compile for speed
@jit
def train_step(params, opt_state, x, y):
    loss_value = loss_fn(params, x, y)
    grads = grad(loss_fn)(params, x, y)
    updates, opt_state = optimizer.update(grads, opt_state)
    params = optax.apply_updates(params, updates)
    return params, opt_state, loss_value

# Train the network
num_epochs = 10000
losses = []

print("Training neural network...")
for epoch in range(num_epochs):
    params, opt_state, loss_value = train_step(params, opt_state, x_norm, y_norm)
    losses.append(loss_value)
    
    if (epoch + 1) % 1000 == 0:
        print(f"Epoch {epoch + 1}/{num_epochs}, Loss: {loss_value:.6f}")

print(f"\nFinal Loss: {losses[-1]:.6f}")

# Make predictions
x_test = jnp.linspace(jnp.min(x_data), jnp.max(x_data), 200).reshape(-1, 1)
x_test_norm = (x_test - x_mean) / x_std
y_pred_norm = forward(params, x_test_norm)
y_pred = y_pred_norm * y_std + y_mean
```
